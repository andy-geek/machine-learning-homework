# Описание задачи

Датасет содержит информацию о более чем 100000 бронирований отелей с июля 2015 года по август 2017 года. Задача заключается в том, чтобы по имеющимся данным определить вероятность отмены брони.

```python
import os
import numpy
import pandas
import matplotlib.pyplot
import seaborn
import sklearn
import sklearn.model_selection
import sklearn.neighbors
import sklearn.ensemble
import sklearn.linear_model
import sklearn.neural_network
```

```python
DATASET_PATH = os.path.abspath(os.path.join(os.getcwd(), 'hotel_booking.csv'))
```

```python
dataset = pandas.read_csv(DATASET_PATH)
dataset.columns = dataset.columns.str.strip()
```

Число объектов и признаков:

```python
dataset.shape
```

# Подготовка датасета

```python
del dataset['reservation_status_date']
del dataset['country']
del dataset['reservation_status']
del dataset['name']
del dataset['email']
del dataset['phone-number']
del dataset['credit_card']
```

Удалим избыточные данные о желаемом времени заселения, оставив данные только о номере недели:

```python
del dataset['arrival_date_year']
del dataset['arrival_date_month']
del dataset['arrival_date_day_of_month']
```

# Обработка пропущенных значений

```python
dataset['meal'].replace({'Undefined': None}, inplace=True)
dataset['market_segment'].replace({'Undefined': None}, inplace=True)
dataset['distribution_channel'].replace({'Undefined': None}, inplace=True)
```

Посмотрим в каких столбцах имеются пропущенные значения:

```python
dataset.isna().sum()[dataset.isna().sum() != 0]
```

Пропуски в признаках children, meal, market_segment и distribution_channel заменим на моду:

```python
dataset['children'].fillna(dataset['children'].mode()[0], inplace=True)
dataset['meal'].fillna(dataset['meal'].mode()[0], inplace=True)
dataset['market_segment'].fillna(dataset['market_segment'].mode()[0], inplace=True)
dataset['distribution_channel'].fillna(dataset['distribution_channel'].mode()[0], inplace=True)
```

Признаки agent и company содержат идентификатор агенства и компании соответственно. В данном случае пропуски можно связать с тем, что бронирующий не был связан с компанией или агентсвом.

```python
dataset['agent'].fillna(0, inplace=True)
dataset['company'].fillna(0, inplace=True)
```

```python
dataset['agent'] = numpy.where(dataset['agent'] > 0, 1, 0)
dataset['company'] = numpy.where(dataset['company'] > 0, 1, 0)
```

# Обработка категориальных признаков

```python
dataset = pandas.get_dummies(dataset)
```

# Визуализация

Изучим основные метрики:

```python
dataset.describe()
```

Можно заметить подозрительные значения, так max у adults составляет 55 человек.
Посмотрим что это за записи:

```python
dataset[dataset['adults'] > 10]
```

Видно, что все эти бронирования были отменены, причём залог не вносился. Следует оставить эти значения как пример изначально крайне подозрительных.

```python
correlation_matrix = dataset.corr()
matplotlib.pyplot.figure(figsize = (48, 48))
seaborn.heatmap(correlation_matrix, square=True, cmap="coolwarm")
```

```python
matplotlib.pyplot.figure(figsize = (6, 6), dpi=100)
matplotlib.pyplot.plot(dataset['arrival_date_week_number'], dataset['stays_in_week_nights'] + dataset['stays_in_weekend_nights'], 'o', alpha = 0.05)
matplotlib.pyplot.xlabel('Неделя прибытия')
matplotlib.pyplot.ylabel('Число ночей')
matplotlib.pyplot.title('Зависимость числа ночей от недели прибытия')
```

# Нормализация признаков

Нормализацию проведём методом стандартизации:

```python
dataset_normalized = (dataset - dataset.mean(axis = 0)) / dataset.std(axis = 0)
dataset_normalized.describe()
```

# Классификация

Разделим данные на обучающую и тестовую выборку с соотношением 3 к 1:

```python
x = dataset_normalized.drop(['is_canceled'], axis=1)
y = dataset['is_canceled']
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.25, random_state=42)
```

## K ближайших соседей

4 соседа:

```python
model = sklearn.neighbors.KNeighborsClassifier(n_neighbors=4)
model.fit(x_train.values, y_train.values)
y_train_pred = model.predict(x_train.values)
y_test_pred = model.predict(x_test.values)
print(f'Ошибки:\nОбучающая: {sklearn.metrics.mean_squared_error(y_train, y_train_pred)}, Тестовая: {sklearn.metrics.mean_squared_error(y_test, y_test_pred)}')
```

```python
confusion_maxtrix = sklearn.metrics.confusion_matrix(y_test, y_test_pred)
confusion_maxtrix_display = sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_maxtrix, display_labels=model.classes_)
confusion_maxtrix_display.plot()
matplotlib.pyplot.show()
```

8 соседей:

```python
model = sklearn.neighbors.KNeighborsClassifier(n_neighbors=8)
model.fit(x_train.values, y_train.values)
y_train_pred = model.predict(x_train.values)
y_test_pred = model.predict(x_test.values)
print(f'Ошибки:\nОбучающая: {sklearn.metrics.mean_squared_error(y_train, y_train_pred)}, Тестовая: {sklearn.metrics.mean_squared_error(y_test, y_test_pred)}')
```

```python
confusion_maxtrix = sklearn.metrics.confusion_matrix(y_test, y_test_pred)
confusion_maxtrix_display = sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_maxtrix, display_labels=model.classes_)
confusion_maxtrix_display.plot()
matplotlib.pyplot.show()
```

Подбор оптимального количества ближайших соседей:

```python
model = sklearn.neighbors.KNeighborsClassifier()
grid = sklearn.model_selection.GridSearchCV(model, param_grid = {'n_neighbors': range(1, 11)}, cv=10)
grid.fit(x_train.values, y_train.values)
print(f"Оптимальное количество соседей: {grid.best_params_['n_neighbors']}")

best_model = grid.best_estimator_
y_train_pred = best_model.predict(x_train.values)
y_test_pred = best_model.predict(x_test.values)
print(f'\nОшибки:\nОбучающая: {sklearn.metrics.mean_squared_error(y_train, y_train_pred)}, Тестовая: {sklearn.metrics.mean_squared_error(y_test, y_test_pred)}')
```

```python
confusion_maxtrix = sklearn.metrics.confusion_matrix(y_test, y_test_pred)
confusion_maxtrix_display = sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_maxtrix, display_labels=best_model.classes_)
confusion_maxtrix_display.plot()
matplotlib.pyplot.show()
```

## Random Forest

```python
model = sklearn.ensemble.RandomForestClassifier()
model.fit(x_train.values, y_train.values)
y_train_pred = model.predict(x_train.values)
y_test_pred = model.predict(x_test.values)
print(f'Ошибки:\nОбучающая: {sklearn.metrics.mean_squared_error(y_train, y_train_pred)}, Тестовая: {sklearn.metrics.mean_squared_error(y_test, y_test_pred)}')
```

```python
confusion_maxtrix = sklearn.metrics.confusion_matrix(y_test, y_test_pred)
confusion_maxtrix_display = sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_maxtrix, display_labels=model.classes_)
confusion_maxtrix_display.plot()
matplotlib.pyplot.show()
```

## Полносвязная нейронная сеть (1 скрытый слой)

```python
model = sklearn.neural_network.MLPClassifier(hidden_layer_sizes=(50), max_iter=1000, random_state=42)
model.fit(x_train.values, y_train.values)
y_train_pred = model.predict(x_train.values)
y_test_pred = model.predict(x_test.values)
print(f'Ошибки:\nОбучающая: {sklearn.metrics.mean_squared_error(y_train, y_train_pred)}, Тестовая: {sklearn.metrics.mean_squared_error(y_test, y_test_pred)}')
```

```python
confusion_maxtrix = sklearn.metrics.confusion_matrix(y_test, y_test_pred)
confusion_maxtrix_display = sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_maxtrix, display_labels=model.classes_)
confusion_maxtrix_display.plot()
matplotlib.pyplot.show()
```

```python
param_grid = {
    'alpha': [0.0001, 0.001, 0.01, 0.1, 1],
    'hidden_layer_sizes': [(50), (60), (70), (80), (90), (100), (150)]
}

model = sklearn.neural_network.MLPClassifier(max_iter=1000, random_state=42)
grid = sklearn.model_selection.GridSearchCV(model, param_grid, cv=5)
grid.fit(x_train.values, y_train.values)

print("Лучшие параметры:", grid.best_params_)
```

Создадим новый MLPClassifier (с подобранными гиперпараметрами)

```python
best_model = grid.best_estimator_
y_train_pred = best_model.predict(x_train.values)
y_test_pred = best_model.predict(x_test.values)
print(f'\nОшибки:\nОбучающая: {sklearn.metrics.mean_squared_error(y_train, y_train_pred)}, Тестовая: {sklearn.metrics.mean_squared_error(y_test, y_test_pred)}')
```

```python
confusion_maxtrix = sklearn.metrics.confusion_matrix(y_test, y_test_pred)
confusion_maxtrix_display = sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_maxtrix, display_labels=best_model.classes_)
confusion_maxtrix_display.plot()
matplotlib.pyplot.show()
```

```python
results = grid.cv_results_
alphas = [param['alpha'] for param in results['params']]
neuron_counts = [param['hidden_layer_sizes'] for param in results['params']]
mean_test_scores = results['mean_test_score']
errors = 1 - mean_test_scores

alphas_neuron_counts_errors = list(zip(alphas, neuron_counts, errors))

neuron_counts_data = {}
for alpha, neuron_count, error in alphas_neuron_counts_errors:
    if neuron_count not in neuron_counts_data:
        neuron_counts_data[neuron_count] = {'alpha': [], 'error': []}
    neuron_counts_data[neuron_count]['alpha'].append(alpha)
    neuron_counts_data[neuron_count]['error'].append(error)

for neuron_count, values in neuron_counts_data.items():
    matplotlib.pyplot.plot(values['alpha'], values['error'], label=f'Neuron Counts: {neuron_count}')

matplotlib.pyplot.xlabel('alpha')
matplotlib.pyplot.ylabel('Ошибка')
matplotlib.pyplot.title('Зависимость ошибки (на тестовом наборе) от alpha')
matplotlib.pyplot.legend()
matplotlib.pyplot.show()
```

```python
alpha_data = {}
for alpha, neuron_count, error in alphas_neuron_counts_errors:
    if alpha not in alpha_data:
        alpha_data[alpha] = {'neuron_count': [], 'error': []}
    alpha_data[alpha]['neuron_count'].append(neuron_count)
    alpha_data[alpha]['error'].append(error)

for alpha, values in alpha_data.items():
    matplotlib.pyplot.plot(values['neuron_count'], values['error'], label=f'Alpha: {alpha}')

matplotlib.pyplot.xlabel('Число нейронов')
matplotlib.pyplot.ylabel('Ошибка')
matplotlib.pyplot.title('Зависимость ошибки (на тестовом наборе) от числа нейронов')
matplotlib.pyplot.legend()
matplotlib.pyplot.show()
```

 ## Промежуточные выводы (по полносвязной нейронной сети)

### По зависимости ошибки на тестовом наборе от alpha (график 1):

Вероятно, при совсем маленьких alpha (0.0001) наблюдается некоторое переобучение почти при любом числе нейронов из нашего набора (кроме 80), поскольку затем значение ошибки стремительно падает.  
Для числа нейронов 100 и 150 оно падает вплоть до alpha = 0.1, а для других значений числа нейронов падение завершается раньше.  
Итак, можно предположить, что при большом числе нейронов и маленьком alpha происходило переобучение.

Далее (для числа нейронов 100 и 150 - сразу после alpha = 0.01, для других чисел нейронов - раньше) при увеличении alpha ошибка только увеличивается (судя по всему, наблюдается недообучение):


### По зависимости ошибки на тестовом наборе от числа нейронов (график 2):

На графике 2 видим, что  
- При alpha = 0.0001 (синим) сначала ошибка ведёт себя немонотонно, но затем (после 100 нейронов) - только увеличивается (возможно, далее при увеличении числа нейронов происходит переобучение)
- При alpha = 0.001 (жёлтым) сначала ошибка постепенно снижается, затем (после 80 нейронов) - увеличивается (видимо, дальше уже переобучение, так как ошибка на тестовых данных стремится к нулю)
- При alpha = 0.01 (зелёным) ошибка почти монотонно снижается до значения нейронов, равного 90, затем слегка повышается, а далее - стремится к нулю (видимо, после 100 нейронов происходит переобучение)
- При alpha = 0.1 (красным) также после 100 нейронов вероятно, происходит переобучение
- При alpha = 1 вероятно, произошло недообучение, поскольку значение ошибки стабильно довольно велико и начинает постепенно снижаться к нулю по мере увеличения числа нейронов (хотя и не слишком стремительно).
